```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# 뉴스 헤드라인 정보를 담을 데이터 프레임
df_news_headline = pd.DataFrame(data = {'headline' : []})

start = 20230101
end = 20231125
query = start

# 2023년 1월 1일부터 11월 25일까지의 뉴스 헤드라인 정보로 데이터 프레임을 만듦
while True:
  if end == query:
    break
  else:
    url = 'https://news.naver.com/main/ranking/popularDay.nhn?rankingType=popular_day&date={query}'
    r = requests.get(url)
    html = r.content
    soup = BeautifulSoup(html, 'html.parser')
    titles_html = soup.select('.rankingnews_list')

    for index, news in enumerate(titles_html):
      title = news.find("a").get_text().strip()
      df_news_headline.loc[len(df_news_headline)] = title

    query+=1

# 헤드라인 정보를 로드하고 csv 파일(news.csv)로 만들어 병합한 뒤 google drive에 저장하여 사용
from google.colab import drive
drive.mount('/content/drive/')

# 모델 분석 시 편리하게 사용하기 위하여 dataframe csv 파일로 저장
df_news_headline.to_csv('/content/drive/MyDrive/report/df_news_headline.csv', encoding='utf-8-sig')

df_news_headline
```

```python
from gensim.models.word2vec import Word2Vec
from konlpy.tag import Okt

# OKT(형태소 분석기)를 사용한 토큰화 작업 + 불용어 제거
okt = Okt()

tokenized_data = []
for sentence in df_news_headline['headline']:
    temp_sentence = okt.morphs(sentence, stem=True) # 토큰화
    temp_sentence = [word for word in temp_sentence if not word in stop_words] # 불용어 제거
    tokenized_data.append(temp_sentence)

tokenized_data
```

```python
# Word2Vec을 이용해 헤드라인 데이터 학습
model = Word2Vec(sentences = tokenized_data, vector_size = 100, window = 5, min_count = 5, workers = 4, sg = 0)
```

```python
# 완성된 임베딩 매트릭스의 크기 확인
model.wv.vectors.shape
```

```python
# 출력 테스트
print(model.wv.most_similar("스페인"))
```

```python
# 출력 테스트
print(model.wv.most_similar("황의조"))
```

```python
# 파일로 저장하기
model.wv.save_word2vec_format('word2vec_format')
```
